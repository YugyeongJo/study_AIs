## 데이터 불러오기

```python
import pyLDAvis # chart로 나타내줌
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
import pandas as pd
df_seoul_120_text = pd.read_csv("https://bit.ly/seoul-120-text-csv")
```

```python
df_seoul_120_text
```

## 자연어 전처리

```python
stopwords = []
```

```python
# 토크나이징 함수 정의
from konlpy.tag import Okt
okt = Okt()

def tokenizer(raw, pos=["Noun","Alpha","Verb","Number"], stopword=stopwords):
    return [
        word for word, tag in okt.pos(
            raw, 
            stem=True    # stemming 바뀌나->바뀌다
            )
            if len(word) > 1 and tag in pos and word not in stopword

    ]
```

```python
# tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfVectorizer = TfidfVectorizer(tokenizer=tokenizer, max_df=0.95, min_df=2)
```

```python
features = tfidfVectorizer.fit_transform(df_seoul_120_text['내용'])
features.toarray()[:2]
```

### 3개로 범주화

```python
latentDirichletAllocation = LatentDirichletAllocation(n_components=3, n_jobs=-1)
```

```python
latentDirichletAllocation.fit(features)
```

```python
vis = pyLDAvis.lda_model.prepare(latentDirichletAllocation, features, tfidfVectorizer)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```
